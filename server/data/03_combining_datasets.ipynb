```python
import pandas as pd

rent_df = pd.read_csv("csvs/Rent_Data_Merged.csv")
amenities_df = pd.read_csv("csvs/amenities_data.csv")
```

```python
rent_df
```

```python
amenities_df
```

```python
print("Rent data columns:", rent_df.columns)
print("Amenities data columns:", amenities_df.columns)
```

```python
# Combine 'block' and 'street_name' in rent_df
rent_df["address"] = rent_df["block"].astype(str).str.strip() + " " + rent_df["street_name"].str.strip()

# Standardise the format in amenities_df ('Address')
amenities_df["address"] = amenities_df["Address"].str.strip()
```

```python
amenities_df = amenities_df.drop(columns=["Address"])
amenities_df = amenities_df.drop_duplicates()
```

```python
rent_df.shape, amenities_df.shape
```

```python
# Left Join by Address
combined_df = pd.merge(rent_df, amenities_df, on="address", how="left")
```

```python
combined_df
```

```python
# Entries with missing amenity information
missing_counts = combined_df[["Nearest MRT Station", "Address Lat"]].isnull().sum()
print("Missing values after merge:\n", missing_counts)
```

```python
combined_df.head(n=5)
```

```python
# removing errant values (some rent approval years are "1")
combined_df = combined_df[combined_df['rent_approval_year'] != 1]

# removing certain redundant/irrelevant columns
combined_df.drop(columns=['1room_sold', '2room_sold', '3room_sold', '4room_sold', '5room_sold', 'exec_sold', 'multigen_sold', 'studio_apartment_sold', '1room_rental', '2room_rental', '3room_rental', 'other_room_rental'], inplace=True)
```
